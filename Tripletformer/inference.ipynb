```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
train_loss = np.load('losses/train_losses_sgrA_982081.npy')
val_loss = np.load('losses/val_losses_sgrA_982081.npy')
epochs = np.arange(1, len(train_loss) +1)
```

```python
plt.plot(epochs, train_loss, alpha=0.5, label='Train')
plt.plot(epochs, val_loss, alpha=0.5, label='Validation')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
